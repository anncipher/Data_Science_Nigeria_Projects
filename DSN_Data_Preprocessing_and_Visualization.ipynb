# Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# To plot line graph

plt.plot ([1,2,3,4,5], [2,4,6,8,10], 'r')
plt.title('My First Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# To specify the size of the figure using method figure() and passing the values as a tuple representing rows and columns

plt.figure(figsize = (15,5))
plt.plot ([1,2,3,4,5], [2,4,6,8,10], 'r', [5,6,7,8,9], [10,12,14,16,18], 'g')
plt.title('Two Joint Graphs')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.arange(0, 10)
print (x)

y = np.sin(x)
print (y)

z = np.cos(x)
print (z)

plt.plot(x,y,z)
plt.xlabel('Numbers between 1 and 10')
plt.ylabel('Functions of the numbers')
plt.title('Sine and Cosine Graphs')
plt.legend(['sine', 'cosine'])
plt.show()
```

```python
# Subplot is used to create smaller multiple plots. It takes 3 arguments: nrows, ncols and index

plt.subplot(1,2,1) # 1 row, 2 columns, index 1
plt.plot([1,2,3,4,5], [2,4,6,8,10], 'r')
plt.title('First')

plt.subplot(1,2,2) # 1 row, 2 columns, index 2
plt.plot([1,3,6,9,12], [3,6,9,12,15], 'g')
plt.title('Second')

plt.subplots_adjust(hspace = 0.4) # For adjusting spaces in between subplots
plt.suptitle('Two Subplots')
plt.show()
```

# Data Preprocessing

Necessary to clean and process data for analysis and training.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Downloads/BL-Flickr-Images-Book.csv')
df.head()
```

```python
df.info()
```

```python
to_drop = ['Edition Statement',
                 'Author',
                 'Contributors',
                 'Corporate Author',
                 'Corporate Contributors',
                 'Former owner',
                 'Engraver',
                 'Issuance type',
                 'Shelfmarks']
df.drop(columns=to_drop, inplace=True, axis=1)
```

```python
df.info()
```

```python
# To know number of missing values in each column in the dataset

df.isnull().sum()
```

```python
# To know the percentage of missing value

df.isnull().sum() * 100 / len(df)
```

# Rename Columns

```python
df = pd.read_csv('Downloads/olympics.csv')
df.head()
```

```python
# The argument header=1 is to start from index 0 as the header and ignore anything above that

df = pd.read_csv('Downloads/olympics.csv', header=1)
df.head()
```

```python
# To rename

new_names = {'Unnamed: 0': 'Country',
             '? Summer': 'Summer olympics',
             '01 !': 'Gold',
             '02 !': 'Silver',
             '03 !': 'Bronze',
             '? Winter': 'Winter Olympics',
             '01 !.1': 'Gold.1',
             '02 !.1': 'Silver.1',
             '03 !.1': 'Bronze.1',
             '? Games': '# Games',
             '01 !.2': 'Gold.2',
             '02 !.2': 'Silver.2',
             '03 !.2': 'Bronze.2'}
```

```python
# Rename header

df.rename(columns = new_names, inplace=True)
df.head()
```
